```python
import sys
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
import math
import random
class SliderDemo(QWidget):
    def __init__(self, parent=None):
        super(SliderDemo, self).__init__(parent)
        self.setWindowTitle("QSlider 例子")
        self.resize(300, 100)
        self.timer=QTimer()
        # self.timer.timeout.connect(self.timerf)
        self.timer.start(200)
        layout = QVBoxLayout()
        self.l1 = QLabel("Hello PyQt5")
        self.l1.setAlignment(Qt.AlignCenter)
        layout.addWidget(self.l1)
        # 水平方向
        self.sl = QSlider(Qt.Horizontal)
        # 设置最小值
        self.sl.setMinimum(10)
        # 设置最大值
        self.sl.setMaximum(50)
        # 步长
        self.sl.setSingleStep(3)
        # 设置当前值
        self.sl.setValue(20)
        # 刻度位置，刻度在下方
        self.sl.setTickPosition(QSlider.TicksBelow)
        # 设置刻度间隔
        self.sl.setTickInterval(5)
        layout.addWidget(self.sl)
        # 连接信号槽
        self.sl.valueChanged.connect(self.valuechange)
        self.setLayout(layout)
        print(self.sl.maximum(),type(self.sl.maximum()))


    def valuechange(self):
        print('current slider value=%s' % self.sl.value())
        size = self.sl.value()
        self.l1.setFont(QFont("Arial", size))

    def timerf(self):
        t=random.randint(10,50)

        self.sl.setValue(t)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    demo = SliderDemo()
    demo.show()
    sys.exit(app.exec_())
```

```python
# coding=utf-8
import sys
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *



class Table(QWidget):
    def __init__(self,parent=None):
        super(Table, self).__init__(parent)
        #设置标题与初始大小
        self.setWindowTitle('QTableView表格复选框案例')
        self.resize(500,300)
        self.tableView=QTableView()
        self.model = QStandardItemModel(self.tableView)
        self.tableView.setEditTriggers(QAbstractItemView.NoEditTriggers)
        self.tableView.setSelectionBehavior(QAbstractItemView.SelectRows)
        self.tableView.setShowGrid(False)
        #设置数据层次结构，4行4列
        self.model=QStandardItemModel(4,5)
        t = QCheckBox(self)
        #设置水平方向四个头标签文本内容
        self.model.setHorizontalHeaderLabels(['','姓名','身份证','地址'])

        for row in range(4):
            item_checked = QStandardItem()
            item_checked.setCheckState(Qt.Checked)
            item_checked.setCheckable(True)
            self.model.setItem(row,0, item_checked)
            for column in range(1,5):
                item=QStandardItem('row %s,column %s'%(row,column))
                #设置每个位置的文本值
                self.model.setItem(row,column,item)

        self.tableView.setModel(self.model)
        self.button=QPushButton(text="button")
        self.button.clicked.connect(self.buttonPush)
        #设置布局
        layout=QVBoxLayout()

        layout.addWidget(self.tableView)
        layout.addWidget(self.button)
        self.setLayout(layout)

    # 获取选中行列、内容
    def buttonPush(self):
        print("in")
        for i in range(self.model.rowCount()):
            if self.model.item(i,0).checkState():
                print(str(i)+": selected")
            else:
                print(str(i)+": not selected")


if __name__ == '__main__':
    app=QApplication(sys.argv)
    table=Table()
    table.show()
    sys.exit(app.exec_())
```
